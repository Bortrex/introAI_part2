```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import normalize
%matplotlib inline
```

```python
nb_maps = 3
maps = ["Small", "Medium", "Large"]
toPlot = ["Computation Time", "Score", "# of Nodes"]
nb_toPlot = len(toPlot)
algo = ["DFS", "BFS", "UCS", "A*"]
nb_algo = len(algo)
dfs = (np.asarray([[0.002, 500, 15],            # small
                            [0.067, 422, 307],  # medium
                            [0.14, 319, 371]])) # large
dfs

bfs = (np.asarray([[0.002, 502, 15],
                             [24.67, 570, 15878],
                             [0.93, 434, 1921]]))
bfs
ucs = (np.asarray([[0.0019, 502, 15],
                             [15.97, 570, 12518],
                             [1.32, 434, 1899]]))

# astar = (np.asarray([[0.0049, 502, 14],
#                              [9.04, 570, 9551],
#                              [1.07, 434, 1133]]))

astar = (np.asarray([[0.0029, 502, 14],
                             [8.11, 570, 7161],
                             [0.70, 434, 813]]))

data = np.asarray([dfs, bfs, ucs, astar])
data[0,1,0]
```

```python
plt.figure(figsize=(16, 10))
plt.title("Running times of different search algorithms", fontsize=23)
plt.xlabel("Search Algorithms",fontsize = 23)
plt.ylabel("Running time (s)",fontsize = 23)

bar_width = .5
bar_positions = np.arange(nb_algo)

for i in range(nb_algo):
    
    plt.bar(i, data[i,2,0], width=bar_width) # small - time

plt.tick_params(axis='both', which='major', labelsize=23)
plt.xticks(bar_positions, ["%s" % i for i in (algo)])
plt.legend(["%s" % i for i in (algo)], fontsize=23)
plt.savefig("large_time.png")
```

```python
plt.figure(figsize=(16, 10))
plt.title("Scores of different search algorithms",fontsize = 23)
plt.xlabel("Search Algorithms",fontsize = 23)
plt.ylabel("Score",fontsize = 23)

bar_width = .5
bar_positions = np.arange(nb_algo)

for i in range(nb_algo):    
    plt.bar(i, data[i,2,1], width=bar_width) # score small

plt.tick_params(axis='both', which='major', labelsize=23)
plt.xticks(bar_positions, ["%s" % i for i in (algo)])
plt.legend(["%s" % i for i in (algo)],fontsize = 23)
plt.savefig("large_score.png")
```

```python
plt.figure(figsize=(16, 10))
plt.title("Number of Nodes of different search algorithms",fontsize = 23)
plt.xlabel("Search Algorithms",fontsize = 23)
plt.ylabel("Nodes",fontsize = 23)

bar_width = .5
bar_positions = np.arange(nb_algo)

for i in range(nb_algo):    
    plt.bar(i, data[i,1,2], width=bar_width) # score small

plt.tick_params(axis='both', which='major', labelsize=23)
plt.xticks(bar_positions, ["%s" % i for i in (algo)])
plt.legend(["%s" % i for i in (algo)],fontsize = 23)
plt.savefig("medium_nodes.png")
```

```python
plt.figure(figsize=(16, 10))
plt.title("A* Search Algorithm over all layouts")
plt.xlabel("Maps")
plt.ylabel("Normalize statistics in relation with the Map")

bar_width = .5
bar_positions = np.arange(nb_maps)*bar_width*(nb_toPlot+1)

for i in range(nb_toPlot):
    plt.bar(bar_positions+bar_width*i, astar[:, i], width=bar_width)

plt.xticks(bar_positions+(nb_algo-1)/2*bar_width, ["%s" % i for i in (maps)])
plt.legend(["%s" % i for i in (toPlot)])
plt.savefig("astarBarPlots.png")
```
